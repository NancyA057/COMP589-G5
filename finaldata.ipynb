```python
#import libraries

import numpy as np
import pandas as pd
import matplotlib as plt
import os
```

```python
# read data from file
earth_data = pd.read_csv("eqcat_CEUS1568-2018.csv")
earth_data.head()
```

```python
# Data Analysis of data set
# Determine if there are any missing values
earth_data.isna().sum()
```

```python
# info on data types
earth_data.info()
```

```python
# Number of columns and rows in data set
earth_data.shape
```

```python
# Combine Data columns (year, month, day) and Time()
#earth_data['Date'] = earth_data[earth_data.columns[4:6]].apply(lambda x :' ', axis = 1)
earth_data['Date'] = earth_data['month'].astype(str) + "/" + earth_data['day'].astype(str) + "/" + earth_data['year'].astype(str)
```

```python
# Combine Time columns ( hour, minute, seconds)
#earth_data['Time'] = earth_data[earth_data.columns[7:9]].apply(lambda x :' '. join(x).astype(int), axis = 1)
earth_data['Time'] = earth_data['hour'].astype(str) + ":" + earth_data['minute'].astype(str) + ":" + earth_data['second'].astype(str)
```

```python
earth_data
```

```python
# Data for analysis
earth_df = earth_data[['Date', 'Time', 'Mw', 'long_degE', 'lat_degN', 'depth_km']]
earth_df.head()
```

```python
# Determine data types
earth_df.info()
```

```python
# Determine the number of zeros in the depth column
print(earth_df['depth_km'].value_counts()[0])
```

```python
# Replace the zero values in the depth column with mean 
earth_df.replace(to_replace =0, value= earth_df['depth_km'].mean(), inplace = True)
earth_df
```
